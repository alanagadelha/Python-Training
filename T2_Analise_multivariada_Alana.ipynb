1) As notas obtidas por 10 estudantes a Análise Matemática e o seu QI (quociente de

Notas =  [8, 14, 18, 10, 6.5, 9, 14, 5.2, 10, 13 ] <br>

Utilize o coeficiente $\rho$ de Spearman e o $R$ de Pearson para verificar se as variáveis estão associadas e qual o seu grau de associação entre elas.


````python
# VAMOS COMEÇAR IMPORTANDO ALGUMAS BIBLIOTECAS!

import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import pearsonr, spearmanr, kendalltau
import pandas as pd
````

#Comentário Alana.#####
##Para calcularmos as correlações, foi interessante buscar as fórmulas originais, para entender as variáveis. Sendo assim, coloquei-as abaixo:
### **Correlação de Pearson**

$$
R =
\frac{\sum_{i=1}^{n}(x_i - \bar{x})(y_i - \bar{y})}
{\sqrt{\sum_{i=1}^{n}(x_i - \bar{x})^2} \cdot
\sqrt{\sum_{i=1}^{n}(y_i - \bar{y})^2}}
$$

onde:

- \( x_i, y_i \) = valores das variáveis  
- \( \bar{x}, \bar{y} \) = médias de \( X \) e \( Y \)  
- \( n \) = número de observações

O coeficiente de correlação de Spearman (ρ) avalia a intensidade e a direção da associação monotônica entre duas variáveis.

Por esse motivo, ele é menos sensível a outliers e não exige que a relação seja linear, apenas monotônica.

---

### **Correlação de Spearman**

$$
\rho = 1 -
\frac{6 \sum_{i=1}^{n} d_i^2}{n(n^2 - 1)}
$$

onde:

- \( d_i = R(x_i) - R(y_i) \)  
- \( R(x_i) \): rank da variável \(X\)  
- \( R(y_i) \): rank da variável \(Y\)  
- \( n \): número de observações  
A correlação de Pearson mede a intensidade e a direção da relação linear entre duas variáveis contínuas.
Ele usa diretamente os valores originais e pressupõe uma tendência linear.

Acima estão as fórmulas originais dos coeficientes, mas para este exercício usarei as funções já criadas no repositório do scipy.

````python
dados = pd.DataFrame({'NOTAS': [8, 14, 18, 10, 6.5, 9, 14, 5.2, 10, 13],
                      'QI': [70, 190, 304, 100, 42, 80, 169, 27, 105, 159]})

# Calculando a correlação de SPEARMAN para o DataFrame
coef_spearman, p_spear = spearmanr(dados['NOTAS'], dados['QI'])

# Calculando a correlação de PEARSON para o DataFrame.
coef_pearson, p_pear = pearsonr(dados['NOTAS'], dados['QI'])

print(f"Spearman (scipy) = {coef_spearman:.4f}  | p-valor = {p_spear:.4e}")
print(f"Pearson  (scipy) = {coef_pearson:.4f}  | p-valor = {p_pear:.4e}")
````

# Comentário Alana:
# Interpretação dos resultados:

 * ρ de Spearman	0.9939	=> Associação monotônica quase perfeita
 * R de Pearson	0.9834	=> Associação linear forte!

Tanto a função **pearsonr** quanto **spearmanr** retornam dois valores. O primeiro valor é o coeficiente de correlação e o segundo é o chamado p-valor, uma medida estatística que informa a chance do resultado da correlação ser fruto do acaso, e não de uma correlação de fato. Neste exercício, tanto **Pearson** quanto **Spearman** indicam uma relação extremamente forte, mas Spearman é ligeiramente maior, indicando que a ordem dos alunos segue quase perfeitamente a ordem de QI.

________________________________________

CONCLUSÂO: As variáveis estão fortemente associadas, tanto do ponto de vista linear (Pearson) quanto monotônico (Spearman). O grau de associação é muito alto nos dois casos, indicando que quanto maior o QI, maior tende a ser a nota obtida.

3) Use a formulação de mínimos quadrados e ajuste os dados a seguir por uma função do tipo: <br>
### **$w(x,y) = a + bx + cy $**

## Faça gráficos para ilustrar o resultado. Lembre-se que a formulação do método dos mínimos quadrados lineares pode ser simplificada a seguir:

$$ d= Gm \\ G^{T}d=G^{T}Gm \\ m_{est}= (G^{T}G)^{-1} G^{T}d \\ $$

Em que $\mathbf{G}$ é a matriz de sensiblidade, ou de derivadas da função hipótese em relação aos parametros, $\mathbf{d}$ é o vetor de dados observados, neste caso o vetor $\mathbf{w}$.

**Lembre-se de verificar as operações matriciais com atenção, para não cometer erros.**  

````python
# Dados
# ============================
x = np.array([5,1,4,2,6,16,13,14,17,15,19,18,20,21,19])
y = np.array([1,3,2,4,6,9,5,9,10,7,8,10,11,9,13])
#z = np.array([1,7,6,7,9,13,8,11,12,11,13,15,14,12,14])
w = np.array([2,26,25,21,20,19,14,10,6,15,19,23,10,7,3])

# Número de pontos
n = len(x)

# matriz G, matriz de sensibilidade
G = np.column_stack((np.ones(n), x, y))

# Mínimos Quadrados
# m = (G^T G)^(-1) G^T d
GTG_inv = np.linalg.inv(G.T @ G)
m = GTG_inv @ G.T @ w          # m = [a, b, c]
a, b, c = m
print(f"a = {a:.4f}")
print(f"b = {b:.4f}")
print(f"c = {c:.4f}")

# Predição
w_pred = G @ m

# Gráficos
plt.figure(figsize=(10,6))
plt.scatter(w, w_pred, color='royalblue', s=80, label='Valores ajustados')
plt.plot([min(w), max(w)], [min(w), max(w)], 'r--', label='1:1 Ajuste ideal')
plt.xlabel("W-observado")
plt.ylabel("W-ajustado")
plt.title("Ajuste por Mínimos Quadrados: W(x,y) = A + Bx + Cy")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
````

* função para aplicar média móvel em um dado 1D:

```
def media_movel(curva,intervalo=100):
  d = list(curva)
    n = len(curva)

    intervalo = round(intervalo/2.0)

    JM = np.zeros(n)
    JM2 = np.zeros(n)
    for j in range(n):
        md = []
        for i in range(j-intervalo,j+intervalo):
            if i < 0:
                i = j-i
            if i >= n-1:
                i = j-i
            md.append(d[i])
        JM[j] = np.mean(md)
        JM2[j] = np.median(md)
    return JM,JM2

```

* Já o gaussiano apresentou comportamento intermediário, tanto na suavização quanto na preservação do sinal.

````python
W = 50  # janela escolhida para executar as demais técnicas!!!!!!!!!!

# Média móvel
df["GR_mean_50"] = df["GR"].rolling(W).mean()

# Mediana móvel
df["GR_median_50"] = df["GR"].rolling(W).median()

# Média ponderada
weights = np.arange(1, W+1)
df["GR_weight_50"] = df["GR"].rolling(W).apply(
    lambda x: np.average(x, weights=weights) if len(x)==W else np.nan,
    raw=True
)

# Filtro Gaussiano
df["GR_gauss_50"] = df["GR"].rolling(W, win_type="gaussian").mean(std=W/4)
````

````python
# Vou tirar os NaN para o cálculo!!!!!!!!!
mask_mean   = ~np.isnan(df["GR_mean_50"])
mask_median = ~np.isnan(df["GR_median_50"])
mask_weight = ~np.isnan(df["GR_weight_50"])
mask_gauss  = ~np.isnan(df["GR_gauss_50"])

# MÉDIA
rmse_mean = np.sqrt(np.mean((df["GR"][mask_mean] - df["GR_mean_50"][mask_mean])**2))
std_mean  = np.std(df["GR_mean_50"][mask_mean])
amp_mean  = df["GR_mean_50"][mask_mean].max() - df["GR_mean_50"][mask_mean].min()

# MEDIANA
rmse_median = np.sqrt(np.mean((df["GR"][mask_median] - df["GR_median_50"][mask_median])**2))
std_median  = np.std(df["GR_median_50"][mask_median])
amp_median  = df["GR_median_50"][mask_median].max() - df["GR_median_50"][mask_median].min()

# PONDERADA
rmse_weight = np.sqrt(np.mean((df["GR"][mask_weight] - df["GR_weight_50"][mask_weight])**2))
std_weight  = np.std(df["GR_weight_50"][mask_weight])
amp_weight  = df["GR_weight_50"][mask_weight].max() - df["GR_weight_50"][mask_weight].min()

# GAUSSIANA
rmse_gauss = np.sqrt(np.mean((df["GR"][mask_gauss] - df["GR_gauss_50"][mask_gauss])**2))
std_gauss  = np.std(df["GR_gauss_50"][mask_gauss])
amp_gauss  = df["GR_gauss_50"][mask_gauss].max() - df["GR_gauss_50"][mask_gauss].min()

# Plotando as Estatísticas
stats_GR_50 = pd.DataFrame({
    "Filtro": ["mean", "median", "weight", "gauss"],
    "RMSE": [rmse_mean, rmse_median, rmse_weight, rmse_gauss],
    "STD":  [std_mean,  std_median,  std_weight,  std_gauss],
    "Amplitude": [amp_mean, amp_median, amp_weight, amp_gauss],
})

stats_GR_50

````
