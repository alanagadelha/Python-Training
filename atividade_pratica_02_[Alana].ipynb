````python
################Copy from the Questions before##########################
import os
import urllib.request
import pandas as pd

url_data3 = 'https://ndownloader.figshare.com/files/12807389/temp-1991-to-1995-months.csv'
filepath3 = '/content/data/temp-1991-to-1995-months.csv'
url_data4 = 'https://ndownloader.figshare.com/files/12807386/temp-1996-to-2000-months.csv'
filepath4 = '/content/data/temp-1996-to-2000-months.csv'

# Set the working directory
print(os.getcwd())
os.chdir('/content')
data_dir = 'data'
if not os.path.exists(data_dir):
  os.makedirs(data_dir)
  print(f"Directory '{data_dir}' created successfully!!!.")
else:
  print(f"Directory '{data_dir}' already created!!!.", "\n")

# Downloading ahn showing the datas
urllib.request.urlretrieve(url_data3, filepath3)
temp_array3 = pd.read_csv(filepath3, delimiter=',')

urllib.request.urlretrieve(url_data4, filepath4)
temp_array4 = pd.read_csv(filepath4, delimiter=',')
#############################################################

# Concat both DataFrames
combined_temp_df = pd.concat([temp_array3, temp_array4], ignore_index=True)
filtered_df = combined_temp_df[(combined_temp_df['Year'] >= 1993) & (combined_temp_df['Year'] <= 1997)]
print("Filtered DataFrame (1993–1997):\n", filtered_df)

# Mean calc by pandas
def column_means(df):
    df_months_only = df.drop(columns='Year')
    return df_months_only.mean()
column_means(filtered_df)
print("\nMean Monthly Temperatures (1993–1997) by Pandas:\n",column_means(filtered_df))

# Mean calc by NumPy
import numpy as np
def column_means(filtered_df):
    filtered_df_without_year = filtered_df.drop(columns='Year')
    return np.mean(filtered_df_without_year, axis=0)
column_means(filtered_df)
print("\nMonthly Temperature (1993–1997) by Numpy \n", column_means(filtered_df))

#################################################################################
####Lets to do the same calc for the Precipitation###############################
#################################################################################
import os
import urllib.request
import pandas as pd
import numpy as np
url_data2 = 'https://ndownloader.figshare.com/files/12807383/onthly-precip-1993-to-1997.csv'
filepath2 = '/content/data/monthly-precip-1993-to-1997.csv'

# Set the working directory
print(os.getcwd())
os.chdir('/content')
data_dir = 'data'
if not os.path.exists(data_dir):
  os.makedirs(data_dir)
  print(f"Directory '{data_dir}' created successfully!!!.")
else:
  print(f"Directory '{data_dir}' already created!!!.", "\n")

urllib.request.urlretrieve(url_data2, filepath2)
precip_array2 = np.loadtxt(filepath2, delimiter=',')

def column_means(precip_array2):
    return np.mean(precip_array2, axis=0)
column_means(precip_array2)

print("########### NEW DATASET ##############################\n")
print("Monthly Precipitation between 1993 to 1997 (inches): \n", column_means(precip_array2))
````
